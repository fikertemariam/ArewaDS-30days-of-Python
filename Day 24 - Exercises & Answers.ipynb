```python
import pandas as pd

# Replace with your file path
file_path = r'C:\Users\Dell\Downloads\names\yob2014.txt'

# Read the file into a Pandas DataFrame
df = pd.read_csv(file_path, header=None, names=['Name', 'Gender', 'Count'])

# Filter the DataFrame for boys in 2014
boys_2014 = df[(df['Gender'] == 'M')]

# Find the third-most popular baby name for boys in 2014
third_most_popular_name = boys_2014.sort_values(by='Count', ascending=False)['Name'].iloc[2]

print(f'The third-most popular baby name for boys in 2014: {third_most_popular_name}')
```
