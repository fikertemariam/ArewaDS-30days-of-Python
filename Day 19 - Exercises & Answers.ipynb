```python
#Read the countries_data.json data file in data directory, create a function that finds the ten most spoken languages
import json

def most_spoken_languages(filename, top_n):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            data = json.load(file)

        languages_count = {}
        for country in data:
            if 'languages' in country:
                for language in country['languages']:
                    languages_count[language] = languages_count.get(language, 0) + 1

        sorted_languages = sorted(languages_count.items(), key=lambda x: x[1], reverse=True)

        return sorted_languages[:top_n]

    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")

# Specify the file path for countries_data.json
json_file_path = r'C:\Users\Dell\Desktop\ArewaDS Workshop\Arewads_30days\DAY 19\countries_data.json'

# Example usage with top 10 languages
result_10 = most_spoken_languages(filename=json_file_path, top_n=10)
print(result_10)

# Example usage with top 3 languages
result_3 = most_spoken_languages(filename=json_file_path, top_n=3)
print(result_3)
```

```python
#4.Extract all incoming email addresses as a list from the email_exchange_big.txt file.
import re

def extract_email_addresses(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()

            # Use a regular expression to find email addresses
            email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')
            email_addresses = re.findall(email_pattern, content)

            return email_addresses

    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")

# Specify the file path for email_exchange_big.txt
txt_file_path = r'C:\Users\Dell\Desktop\ArewaDS Workshop\Arewads_30days\DAY 19\email_exchange_big.txt'

# Extract email addresses and print the result
email_addresses = extract_email_addresses(txt_file_path)
print(email_addresses)
```

```python
import re
from collections import Counter

def find_most_common_words(input_data, num_words):
    # If input_data is a file path, read the content from the file
    if isinstance(input_data, str) and input_data.endswith('.txt'):
        try:
            with open(input_data, 'r', encoding='utf-8') as file:
                content = file.read()
        except FileNotFoundError:
            print(f"Error: File '{input_data}' not found.")
            return []
    elif isinstance(input_data, str):
        content = input_data
    else:
        print("Error: Invalid input data.")
        return []

    # Use regex to extract words from the content
    words = re.findall(r'\b\w+\b', content.lower())

    # Count the occurrences of each word
    word_counts = Counter(words)

    # Get the most common words
    most_common_words = [(count, word) for word, count in word_counts.items()]

    # Sort the list in descending order by count
    most_common_words.sort(reverse=True)

    return most_common_words[:num_words]

# Specify the file path for 'sample.txt'
file_path = r'C:\Users\Dell\Desktop\ArewaDS Workshop\Arewads_30days\DAY 19\sample.txt'

# Example usage with the top 10 most common words
result_10 = find_most_common_words(file_path, 10)
print(result_10)

# Example usage with the top 5 most common words
result_5 = find_most_common_words(file_path, 5)
print(result_5)
```

```python
import re
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def clean_text(text):
    # Remove special characters, numbers, and convert to lowercase
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
    return cleaned_text.lower()

def remove_support_words(text, stop_words):
    # Tokenize the text and remove stop words
    words = re.findall(r'\b\w+\b', text)
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)

def check_text_similarity(text1, text2, stop_words):
    # Clean and remove support words from the texts
    cleaned_text1 = remove_support_words(clean_text(text1), stop_words)
    cleaned_text2 = remove_support_words(clean_text(text2), stop_words)

    # Use TF-IDF vectorizer to convert text to vectors
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([cleaned_text1, cleaned_text2])

    # Calculate cosine similarity between the vectors
    similarity = cosine_similarity(vectors[0], vectors[1])[0][0]

    return similarity

def read_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None

# Specify the file paths
michelle_speech_path = r'C:\Users\Dell\Desktop\ArewaDS Workshop\Arewads_30days\DAY 19\michelle_obama_speech.txt'
melania_speech_path = r'C:\Users\Dell\Desktop\ArewaDS Workshop\Arewads_30days\DAY 19\melina_trump_speech.txt'
stop_words_path = r'C:\Users\Dell\Desktop\ArewaDS Workshop\Arewads_30days\DAY 19\stop_words.py'

# Read the text content and stop words
michelle_text = read_file(michelle_speech_path)
melania_text = read_file(melania_speech_path)
stop_words = read_file(stop_words_path).splitlines()

# Check the similarity between Michelle's and Melania's speeches
similarity_score = check_text_similarity(michelle_text, melania_text, stop_words)
print(f"Similarity Score: {similarity_score}")
```

```python
#8.Find the 10 most repeated words in the romeo_and_juliet.txt
import re
from collections import Counter

def find_most_repeated_words(file_path, num_words):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None

    # Use regex to extract words from the content
    words = re.findall(r'\b\w+\b', content.lower())

    # Count the occurrences of each word
    word_counts = Counter(words)

    # Get the most repeated words
    most_repeated_words = word_counts.most_common(num_words)

    return most_repeated_words

# Specify the file path for romeo_and_juliet.txt
file_path = r'C:\Users\Dell\Desktop\ArewaDS Workshop\Arewads_30days\DAY 19\romeo_and_juliet.txt'

# Find the 10 most repeated words
most_repeated_words = find_most_repeated_words(file_path, 10)

# Print the result
print(most_repeated_words)

```

```python
#9.Read the hacker news csv file and find out: a) Count the number of lines containing python or Python b) Count the number lines containing JavaScript, javascript or Javascript c) Count the number lines containing Java and not JavaScript
import csv

def count_lines_with_keyword(file_path, keyword):
    try:
        with open(file_path, 'r', encoding='utf-8', newline='') as file:
            reader = csv.reader(file)
            lines_count = sum(1 for row in reader if keyword.lower() in row[1].lower())
        return lines_count
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None

# Specify the file path for hacker_news.csv
file_path = r'C:\Users\Dell\Desktop\ArewaDS Workshop\Arewads_30days\DAY 19\hacker_news.csv'

# Count the number of lines containing Python or python
python_lines_count = count_lines_with_keyword(file_path, 'python')
print(f"a) Number of lines containing Python or python: {python_lines_count}")

# Count the number of lines containing JavaScript, javascript, or Javascript
javascript_lines_count = count_lines_with_keyword(file_path, 'javascript')
print(f"b) Number of lines containing JavaScript, javascript, or Javascript: {javascript_lines_count}")

# Count the number of lines containing Java and not JavaScript
java_lines_count = count_lines_with_keyword(file_path, 'java')
javascript_lines_count = count_lines_with_keyword(file_path, 'javascript')
java_not_javascript_lines_count = java_lines_count - javascript_lines_count
print(f"c) Number of lines containing Java and not JavaScript: {java_not_javascript_lines_count}")
```
